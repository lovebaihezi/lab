```
{tessella, lena} = Import /@ {"./Tessella.bmp", "./Lena.bmp"};
noise = {"GaussianNoise", #}& /@ {0, 0.02};
```

```
AddNoise[image_] := ImageEffect[image, #]& /@ noise;
```

```
{tessella, lena} = Map[AddNoise, {tessella, lena}];
```

```
EdgeDetect[Part[tessella, 1]]
EdgeDetect[Part[tessella, 2]]
```

```
EdgeDetect[Part[lena, 1], #]& /@ Range[0, 10, 0.5]
EdgeDetect[Part[lena, 2], #]& /@ Range[0, 10, 0.5]
```
