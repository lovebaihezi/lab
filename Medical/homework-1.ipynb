```python
# COLAB specific grammar for install package
!pip install pydicom matplotlib
import pydicom as dicom
import matplotlib.pyplot as plt
```

```python
file_path: str = "./"
file_name: str = "0004.DCM"
dicom_file = dicom.dcmread(file_path + file_name)
print(dicom_file)
```

```python
print(dicom_file.pixel_array.shape)
byte_array = dicom_file.pixel_array[:, 128:(512 - 128), 128:(512 - 128)]
for i in range(17):
  plt.subplot(4, 5, i+1)
  plt.imshow(byte_array[i])
```
